```python
import pandas as pd     #імпорт lib pandas
import matplotlib.pyplot as plt     #імпорт pyplon з lib matplotlib (необхідна для побудови графіків)     


column_names = ['times_pregnant', 'plasma_glucose', 'blood_pressure','triceps_skin', 'serum_insulin', 'body_mass_index','diabetes_pedigree_function', 'age', 'class_variable']      #list що містить назву стовпців
url ="https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"       #адреса з якої буде завантажений набір даних 
dataset = pd.read_csv(url, names=column_names)      #використовуємо список для надання назв стовпяів які завантажуються у DataFrame

missing_data = dataset.isna().sum() #insa визначає пропущені значення а sum обчислює загальну кількість пропущених значень пісял чого результат зберігається у missing_data

print(missing_data, "\n")
```

```python
import pandas as pd     #імпорт lib pandas
import numpy as np      #імпорт lib numpy
import matplotlib.pyplot as plt

url ="https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"       #адреса з якої буде завантажений набір даних 

columns_to_replace_zero = ['times_pregnant', 'plasma_glucose', 'blood_pressure','triceps_skin', 'serum_insulin', 'body_mass_index','diabetes_pedigree_function', 'age','class_variable']       #list що містить назву стовпців  

# Завантаження даних з CSV-файлу за вказаною URL-адресою у DataFrame info
# Визначення назв стовпців за допомогою списку columns_to_replace_zero
info = pd.read_csv(url, names=columns_to_replace_zero)

# Виведення середнього значення кожного стовпця у DataFrame info
print("Average:", "\n", info.mean(), "\n")

# Виведення дисперсії для кожного стовпця у DataFrame info
print("Variance:" , "\n", info.var(), "\n")

# Виведення стандартного відхилення для кожного стовпця у DataFrame info
print("Standard Deviation:", "\n", info.std(), "\n")


# Побудова гістограм
for column in column_names:
    plt.figure(figsize=(8, 6))
    plt.hist(dataset[column].dropna(),  color='black')
    plt.title(f'{column}')
    plt.grid(True)
    plt.show()
```
